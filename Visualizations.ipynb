```python
df1 = pd.read_csv('all_time_payment_type_usage.csv')
df1
```

```python
import plotly.express as px

df1 = df1[df1['no_usage']>3]
# Create plot
fig_reviews = px.pie(df1, values = 'no_usage', color = 'payment_type',
                     names = 'payment_type', color_discrete_sequence = ['#8DCEB6','#F48484', '#F5EA5A', '#C780FA'],
                     hole = 0.6)



# Update traces
fig_reviews.update_traces(textposition = 'outside', textfont = dict(color = 'dark blue', size = 15), textinfo = 'label+percent', pull = [0,0,0,0.2], rotation = 200)

# Add Annotations
fig_reviews.add_annotation (text = '<b>Payment Type\nUsage Share</b>', showarrow = False, font = dict(size = 12, color='dark blue'))

# Show Plot
fig_reviews.show()
```
